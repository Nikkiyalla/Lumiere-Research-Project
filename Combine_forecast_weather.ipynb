```python
df_weather_period = pd.DataFrame()
datetime_format = '%Y-%m-%d %H:%M:%S'

for site_number in range(1,5):
  site_name = sites[site_number]['Name'].lower()

  reg_exp = "^(forecast_{})".format(site_name)

  for filename in os.listdir(input_data_path):
    #print(filename)
    if re.match(reg_exp, filename):
      #get datetime when the forecast was made
      #This info is stored in the filename
      date_str = filename.split('_')[2]
      time_str = filename.split('_')[3]
      time_str = time_str.replace("-", ":")

      datetime_str = date_str + ' ' + time_str

      datetime_obj = datetime.strptime(datetime_str, datetime_format)

      df_weather_day = pd.read_csv(input_data_path.joinpath(filename), parse_dates=['dt'], index_col=0)
      df_weather_day.insert(1, 'city', site_name)
      df_weather_day.insert(1, 'dt_forecast_made', datetime_obj)
      df_weather_period = pd.concat([df_weather_period, df_weather_day])

df_weather_period.reset_index(inplace= True)
#print(filename)
df_weather_period
```

```python
df_weather_period = pd.DataFrame()
datetime_format = '%m-%d-%Y %H:%M:%S'

for site_number in range(1,5):
  site_name = sites[site_number]['Name']

  reg_exp = "^(current_forecast_{})".format(site_name)

  for filename in os.listdir(input_data_path):
    #print(filename)
    if re.match(reg_exp, filename):
      #get datetime when the forecast was made
      #This info is stored in the filename
      date_str = filename.split('_')[3]
      time_str = filename.split('_')[4]
      time_str = time_str.replace("-", ":")

      datetime_str = date_str + ' ' + time_str + ':00'

      datetime_obj = datetime.strptime(datetime_str, datetime_format)

      df_weather_day = pd.read_csv(input_data_path.joinpath(filename), parse_dates=['dt'], index_col=0)
      df_weather_day.insert(1, 'city', site_name)
      df_weather_day.insert(1, 'dt_forecast_made', datetime_obj)
      df_weather_period = pd.concat([df_weather_period, df_weather_day])

df_weather_period.reset_index(inplace= True)
df_weather_period_02 = df_weather_period.copy()
```

```python
#combining both dataframes
df_weather_period_both = pd.concat([df_weather_period_02, df_weather_period_01], axis =0)
df_weather_period_both
```

```python
df_weather_period_both.info()
```

```python
# Add number of hours difference between time where forecast was made
# and time of forecast
def calculate_hours_diff(datetime_forec_made, datetime_forecast):
  datetime_forec_made = datetime_forec_made.replace(minute=0)
  datetime_diff = datetime_forecast - datetime_forec_made
  days, seconds = datetime_diff.days, datetime_diff.seconds
  hours = days * 24 + seconds // 3600
  return hours

df_weather_period_both.insert(2, 'dt_hours_diff','')
df_weather_period_both['dt_hours_diff'] = df_weather_period_both.apply(lambda x: calculate_hours_diff(x['dt_forecast_made'],x['dt']), axis=1)

df_weather_period_both
```

```python
#sort values by datetime difference and eliminate duplicates based on dt

df_weather_period_both.sort_values(by='dt_hours_diff', ascending=True, inplace = True)
df_weather_period_both.drop_duplicates(subset=['dt', 'city'], keep='first', inplace = True)
```

```python
#Save dataframe in data/inputs folder
df_weather_period_both.sort_values(by=['dt', 'city'], ascending=True, inplace = True)
out_filename = 'weather_forecast_multicities.csv'
df_weather_period_both.to_csv(input_data_path.joinpath(out_filename))
```
